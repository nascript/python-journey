```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

## 1. GENERATE DATA


```python
jumlah_data = 100

x = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
y = np.array([i*0.1 for i in range(jumlah_data)])
```

## 2. PLOT DATA


```python
# plt.scatter(x,y)
# plt.show()
```

## 3. FUNGSI LINEAR


```python
def linerFunc(x,gradien):
  y = gradien*x
  print(y)
  return y
```

### 3.1 menghitung persamaan linear awal


```python
x_prediction = np.array([0, 10])
m_init = 5
y_prediction = linerFunc(x_prediction,m_init)
print(y_prediction)
```

```python
# plt.scatter(x,y)
# plt.plot(x_prediction,y_prediction,"red")
# plt.axis([0,10,0,10])
# plt.show()
```

```python
m_prediciton = m_init

m_list_prediction = []
y_list_prediction = []
x_list_prediction = []
learning_rate = 0.1

for i in range(1,jumlah_data):
  y_prediction = linerFunc(x[i], m_prediciton)
  y_actual = y[i]

  error = y_actual - y_prediction
  delta_m = learning_rate * error / x[i]

  m_prediciton = m_prediciton + delta_m

  m_list_prediction.append(m_prediciton)
  y_list_prediction.append(y_prediction)
  x_list_prediction.append(x[i])
```

### 3.2 Visualisasi Prediksi


```python

fig = plt.figure(figsize=(4,4))
line, = plt.plot([],[],"Red")


def animate(frame_num):
    if 0 <= frame_num < len(x_list_prediction):
        x = [0, x_list_prediction[frame_num]]
        y = [0, y_list_prediction[frame_num]]
        line.set_data((x, y))
        return line
    else:
        # Jika frame_num di luar rentang, hentikan animasi
        anime.event_source.stop()


plt.scatter(x,y)
plt.axis([0,10,0,10])
anime = FuncAnimation(fig,animate,frames=100,interval=100,repeat=False)
plt.show()
```
